Data Types Issue:

Date Column: Date is currently object, should be converted to datetime.

# 2. Data Pre-processing Steps

# Remove extra spaces in column names

```python
df_financials.columns
```

```python
df_financials.columns = df_financials.columns.str.strip()
```

# Convert numeric columns to proper datatypes

```python
numeric_cols = [
    "Units Sold", "Manufacturing Price", "Sale Price", 
    "Gross Sales", "Discounts", "Sales", "COGS", "Profit"
]

for col in numeric_cols:
    df_financials[col] = (
        df_financials[col]
        .astype(str)            
        .str.replace(" ", "")    
        .str.replace("$", "")    
        .str.replace(",", "")
        .str.replace("$-","")
        .str.replace("$- ","")
    )
    df_financials[col] = pd.to_numeric(df_financials[col], errors="coerce")
```

```python
df_financials.dtypes
```

# Convert Date column to datetime 

```python
df_financials["Date"] = pd.to_datetime(df_financials["Date"])
```

```python
df_financials.dtypes
```

# Create derived columns (example: Profit Margin %)

```python
df_financials["Profit Margin %"] = (df_financials["Profit"] / df_financials["Sales"]) * 100
```

```python
df_financials['Profit Margin %'] = df_financials['Profit Margin %'].fillna((df_financials['Profit'] / df_financials['Sales']) * 100)
```

```python
print(df_financials[["Sales", "Profit", "Profit Margin %"]].head())
```

# Aggregation example (Total Sales by Year & Segment)

```python
agg_sales = df_financials.groupby(["Year", "Segment"])["Sales"].sum().reset_index()
```

```python
agg_sales.head()
```

# After Data Pre-processing:

Correct numeric columns.

Proper datetime column.

A new derived feature: Profit Margin %.

Aggregated sales summary for analysis.

```python
df_financials.isnull().sum()
```

```python
df_financials['Discounts'] = df_financials['Discounts'].fillna(0)
df_financials['Profit'] = df_financials['Profit'].fillna(0)
df_financials['Profit Margin %'] = df_financials['Profit Margin %'].fillna(0)
```
